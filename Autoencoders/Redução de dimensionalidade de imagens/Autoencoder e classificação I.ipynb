### Autoencoder e classificação I

**Importações iniciais**

```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import mnist
from keras.models import Model, Sequential
from keras.layers import Input, Dense
from keras.utils import np_utils
```

**Leitura da base de dados**

```python
(previsores_train, classe_train), (previsores_test, classe_test) = mnist.load_data()
```

**Normalização dos dados**

```python
previsores_train = previsores_train.astype('float32') / 255
previsores_test = previsores_test.astype('float32') / 255
classe_dummy_train = np_utils.to_categorical(classe_train)
classe_dummy_test = np_utils.to_categorical(classe_test)
```

**Modificando a forma dos nossos dados**

```python
previsores_train = previsores_train.reshape(len(previsores_train), np.prod(previsores_train.shape[1:]))
```

```python
previsores_test = previsores_test.reshape(len(previsores_test), np.prod(previsores_test.shape[1:]))
```

**Criação do Autoencoder**

```python
autoencoder = Sequential()
autoencoder.add(Dense(units = 32, activation = 'relu', input_dim = 784))
autoencoder.add(Dense(units = 784, activation = 'sigmoid'))
```

**Compilação do Autoencoder**

```python
autoencoder.compile(optimizer = 'adam', loss = 'binary_crossentropy',
                    metrics = ['accuracy'])
```

```python
autoencoder.fit(previsores_train, previsores_train,
               epochs = 50, batch_size = 256,
               validation_data = (previsores_test, previsores_test))
```

**Visualização das imagens**

```python
dimensao_original = Input(shape=(784,))
camada_encoder = autoencoder.layers[0]
```

```python
encoder = Model(dimensao_original, camada_encoder(dimensao_original))
```

```python
previsores_train_cod = encoder.predict(previsores_train)
previsores_test_cod = encoder.predict(previsores_test)
```

**Criação das Redes Neurais**

Sem redução de dimensionalidade:

```python
classificador1 = Sequential()
classificador1.add(Dense(units = 397, activation = 'relu', input_dim = 784))
classificador1.add(Dense(units = 397, activation = 'relu'))
classificador1.add(Dense(units = 10, activation = 'softmax'))

classificador1.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['accuracy'])

classificador1.fit(previsores_train, classe_dummy_train, batch_size = 256,
                  epochs = 100, validation_data = (previsores_test, classe_dummy_test))
```

Com redução de dimensionalidade:

```python
classificador2 = Sequential()
classificador2.add(Dense(units = 21, activation = 'relu', input_dim = 32))
classificador2.add(Dense(units = 21, activation = 'relu'))
classificador2.add(Dense(units = 10, activation = 'softmax'))

classificador2.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
                      metrics = ['accuracy'])

classificador2.fit(previsores_train_cod, classe_dummy_train, batch_size = 256,
                  epochs = 100, validation_data = (previsores_test_cod, classe_dummy_test))
```
